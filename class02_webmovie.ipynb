### 영화 리뷰 정보 가져오기

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
# url = 'https://movie.naver.com/movie/point/af/list.nhn?st=mcode&sword=171725&target=after'
basic_url = "https://movie.naver.com/movie/point/af/list.nhn?st=mcode&sword=186613&target=after&page="
# 2page 
```

```python
basic_url + str("1")
```

```python
url = basic_url + str("1")
page = urlopen(url)  # html 정보 가져오기 
soup = BeautifulSoup(page, "lxml")  # html을 파싱 또는 구조화.

comment_all = soup.find_all('td', class_='title')
a = list(comment_all[2].children)
a
```

```python
comments = [ ]
cnt = 1
for i in range(1,50):
    url = basic_url + str(i)
    page = urlopen(url)
    soup = BeautifulSoup(page, "lxml")

    comment_all = soup.find_all('td', class_='title')
    for comment in comment_all:
        temp= list(comment.children)
        if len(temp) < 7:
            cnt= cnt + 1
            continue
        else:
            result = temp[6].strip()
            comments.append(result)
print(len(comments))
print(comments)
print(cnt)
```

```python
import pandas as pd
```

```python
dict_doc = {"text" : comments}
doc = pd.DataFrame(dict_doc)
```

```python
doc.to_csv("작은아씨들.csv", index = False)
```
