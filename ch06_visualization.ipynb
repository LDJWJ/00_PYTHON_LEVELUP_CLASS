### 6-3 항공기 승객에 대한 데이터 시각화

```
import seaborn as sns
```

```
fl_data = sns.load_dataset("flights")
fl_data
```

```
sns.barplot(x="year", y="passengers", data=fl_data)
```

```
sns.barplot(x="year", y="passengers", data=fl_data)
plt.title("Passengers by Year")
plt.xlabel("year(1949~1960)")
plt.ylabel("flights passengers")
plt.show()
```

### 실습해 보기

```
sns.barplot(x="month", y="passengers", data=fl_data)
plt.title("Passengers by Year")
plt.xlabel("year(1949~1960)")
plt.ylabel("flights passengers")
plt.show()
```

```
sns.barplot(x="passengers", y="month", data=fl_data)
plt.title("Passengers by Year")
plt.xlabel("year(1949~1960)")
plt.ylabel("flights passengers")
plt.show()
```

### 6-5 지도 정보 시각화 – folium 입문 

```
import folium
```

```
# 버전 확인
print(folium.__version__)
```

```
# 위도(latitude), 경도(longitude) 를 이용한 위치표시
stamen = folium.Map(location=[37.5838699,127.0565831])
stamen
```

```
# 위치 표시, 확대 (zoom_start 값이 크면 확대큼.)
# default = 10, zoom_start
stamen = folium.Map(location=[37.5838699,127.0565831], 
              zoom_start=15)
stamen
```

```
## 원모양의 마커
## 서울시립대학교 우체국 마커
## folium.Marker
## folium.CircleMaker
## 구글 지도 이용하여 위도, 경도 확인하기
stamen = folium.Map(location=[37.583891,127.058784], zoom_start=25)

folium.Marker([37.583000, 127.059771]).add_to(stamen) # 서울 시립 대학교 대강당
folium.Marker([37.583813, 127.057893]).add_to(stamen) # 서울 시립 대학교 건설 공학관
stamen
```
